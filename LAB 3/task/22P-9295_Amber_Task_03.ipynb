```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

columns = [ 'SeniorCitizen', 'tenure', 'MonthlyCharges']

df_filtered = df.copy()  
for col in columns:
    df_filtered = remove_outliers_iqr(df_filtered, col)

print(f"Original data shape: {df.shape}")
print(f"Data shape after removing outliers: {df_filtered.shape}")

df_cleaned = df_filtered.copy()

plt.figure(figsize=(15, 10))

for i, col in enumerate(columns):
    plt.subplot(len(columns), 2, 2 * i + 1)
    sns.boxplot(x=df[col])
    plt.title(f'Before Outlier Removal: {col}')

    plt.subplot(len(columns), 2, 2 * i + 2)
    sns.boxplot(x=df_cleaned[col])
    plt.title(f'After Outlier Removal: {col}')

plt.tight_layout()
plt.suptitle('Outlier Removal for Multiple Columns', fontsize=18, y=1.02)

plt.show()
```
