```python
import matplotlib.pyplot as plt
import seaborn as sns

# Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

```python

categorical_features = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'MultipleLines',
    'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 
    'PaperlessBilling', 'PaymentMethod'
]

plt.figure(figsize=(15, 20))

for i, feature in enumerate(categorical_features):
    plt.subplot(5, 3, i + 1)  
    sns.countplot(data=df, x=feature, hue='Churn')
    plt.title(f'{feature} vs Churn')
    plt.xticks(rotation=45)
    plt.xlabel(feature)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
from sklearn.impute import SimpleImputer
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
imputer = SimpleImputer(strategy='mean')
df[numerical_cols] = imputer.fit_transform(df[numerical_cols])
print(df.head())
```

```python
df.drop_duplicates(inplace=True)
```

```python
import numpy as np
df.replace('-', np.nan, inplace=True)
```

```python

from sklearn.preprocessing import OneHotEncoder
one_hot_encoder= OneHotEncoder(sparse_output=False)
MultipleLines = df['MultipleLines'].values.reshape(-1, 1)
df['MultipleLines']= one_hot_encoder.fit_transform(MultipleLines)

Contract = df['Contract'].values.reshape(-1, 1)
df['Contract']= one_hot_encoder.fit_transform(Contract)

PaymentMethod = df['PaymentMethod'].values.reshape(-1, 1)
df['PaymentMethod']= one_hot_encoder.fit_transform(PaymentMethod)
```

Why I used this encoding??

I have used One-Hot Encoding for the features MultipleLines, Contract, and PaymentMethod because these features have nominal categories with no inherent order or ranking. One-Hot Encoding ensures that the model does not infer any false ordinal relationships between categories.

MultipleLines: Categories like 'No', 'Yes', and 'No phone service' have no meaningful order, so using One-Hot Encoding avoids assigning numeric values that could imply one is greater than the other.

Contract: The values 'Month-to-month', 'One year', and 'Two year' are nominal, and although they may appear ordinal, their relationship is not linear or numeric. 

PaymentMethod: This feature includes different payment methods like 'Electronic check', 'Mailed check', 'Bank transfer', etc., which are nominal categories. One-Hot Encoding correctly treats them as distinct categories with no inherent order.

```python
from sklearn.preprocessing import LabelEncoder
label_encoder= LabelEncoder()
df['gender'] = label_encoder.fit_transform(df['gender'])
df['Churn'] = label_encoder.fit_transform(df['Churn'])
df['Partner'] = label_encoder.fit_transform(df['Partner'])
df['Dependents'] = label_encoder.fit_transform(df['Dependents'])
df['PhoneService'] = label_encoder.fit_transform(df['PhoneService'])
df['InternetService'] = label_encoder.fit_transform(df['InternetService'])
df['OnlineSecurity'] = label_encoder.fit_transform(df['OnlineSecurity'])
df['OnlineBackup'] = label_encoder.fit_transform(df['OnlineBackup'])
df['DeviceProtection'] = label_encoder.fit_transform(df['DeviceProtection'])
df['TechSupport'] = label_encoder.fit_transform(df['TechSupport'])
df['StreamingTV'] = label_encoder.fit_transform(df['StreamingTV'])
df['StreamingMovies'] = label_encoder.fit_transform(df['StreamingMovies'])
df['PaperlessBilling'] = label_encoder.fit_transform(df['PaperlessBilling'])
df
```

Why I used this encoding?

 Many of the categorical columns have binary categories (e.g., Yes/No, Male/Female). Label Encoding efficiently transforms these binary values into integers (0, 1) without implying any ordinal relationship. Itâ€™s a memory-efficient solution that works well for binary or ordinal features.

```python
df['MonthlyCharges'] = df['MonthlyCharges'].astype(int)
df['tenure'] = df['tenure'].astype(int)
df['Contract'] = df['Contract'].astype(int)
df['TotalCharges'] = label_encoder.fit_transform(df['TotalCharges'])
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
IQR = q3 - q1
print("IQR= ", IQR)

lowerBound = q1 - 1.5 * IQR
upperBound = q3 + 1.5 * IQR
df = df[~((df < lowerBound) | (df > upperBound)).any(axis=1)]

print("DataFrame after removing outliers:", df)
```

```python

plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```
